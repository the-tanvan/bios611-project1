To add on to the writeup produced in R, I did an analysis to see what percentage of Airbnb in Asheville have my desired amenities. The top three priorities for my fiance and me is to have wifi, a stove to cook on, and a hot tub to relax in. Below, I show the percentages of listings with those amenities, as well as the percentage of listings with all three of those desirable traits.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("derived_data/listings.csv")
row_total = df.shape[0]

df.loc[df['amenities'].str.contains('Wifi'),'wifi'] = 'True'
wifi_count=df.wifi.value_counts()

df.loc[df['amenities'].str.contains('Stove'),'stove'] = 'True'
stove_count=df.stove.value_counts()

df.loc[df['amenities'].str.contains('Hot tub'),'hottub'] = 'True'
hottub_count=df.hottub.value_counts()

df.loc[(df['wifi']=="True") & (df['stove']=="True") & (df['hottub']=="True"),'ideal'] = 'True'
ideal_count=df.ideal.value_counts()
```

```python
labels="Wifi","No wifi"
sizes=np.array([wifi_count,row_total-wifi_count]).flatten()
colors = ['#FF585D','#FFD6D8']

fig1,ax1=plt.subplots()
ax1.pie(sizes,labels=labels,colors=colors,autopct='%1.1f%%')
ax1.axis('equal')
plt.title('Pie Chart of Airbnb Listings with Wifi')
plt.show()
```

```python
labels="Stove","No stove"
sizes=np.array([stove_count,row_total-stove_count]).flatten()
colors = ['#FF585D','#FFD6D8']

fig1,ax1=plt.subplots()
ax1.pie(sizes,labels=labels,colors=colors,autopct='%1.1f%%')
ax1.axis('equal')
plt.title('Pie Chart of Airbnb Listings with a Stove')
plt.show()
```

```python
labels="Hot tub","No hot tub"
sizes=np.array([hottub_count,row_total-hottub_count]).flatten()
colors = ['#FF585D','#FFD6D8']

fig1,ax1=plt.subplots()
ax1.pie(sizes,labels=labels,colors=colors,autopct='%1.1f%%')
ax1.axis('equal')
plt.title('Pie Chart of Airbnb Listings with a Hot Tub')
plt.show()
```

```python
labels="Ideal","Not ideal"
sizes=np.array([ideal_count,row_total-ideal_count]).flatten()
colors = ['#FF585D','#FFD6D8']

fig1,ax1=plt.subplots()
ax1.pie(sizes,labels=labels,colors=colors,autopct='%1.1f%%')
ax1.axis('equal')
plt.title('Pie Chart of Airbnb Listings with our Ideal Characteristics')
plt.show()
```
